```python
import  pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
area = [2600,3000,3200,3600,4000]
prices = [550000,565000,610000,680000,725000]
df = pd.DataFrame({"area":area, "prices":prices})
df.head()
```

```python
%matplotlib inline
plt.title("House Pricing - (Area vs Price)")
plt.xlabel("Area (sqt)")
plt.ylabel("Price ($)")
plt.scatter(df["area"],df["prices"],marker="*")
plt.show()
```

```python
model = linear_model.LinearRegression()
model.fit(df[['area']],df["prices"])
```

```python
model.predict([[3300]])   # Hence we are working in single feature we need to add list [features]
```

```python
#y  = m*x + c

c = model.coef_
m = model.intercept_
print(c,m)
```

```python
pre_df = pd.DataFrame([1000,1500,2300,3540,4120,4560,5490,3460,4750,2300,9000,8600,7100], columns=["area"])
pre_df
```

```python
pre_price = model.predict(pre_df)
pre_df["price"] = pre_price
pre_df
```

```python
%matplotlib inline
plt.title("House Pricing - (Area vs Price)")
plt.xlabel("Area (sqt)")
plt.ylabel("Price ($)")
plt.scatter(df["area"],df["prices"],marker="*")
plt.plot(df["area"],model.predict(df[["area"]]),c = "red")
plt.show()
```
